```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
##downloading the home page of the website
url = "https://lk.spaceylon.com/"

headers = {
    "User-Agent": "Mozilla/5.0"
}

response = requests.get(url, headers=headers)

print(response.status_code)
```

```python
###saving the HTML
with open("homepage.html", "w", encoding="utf-8") as file:
    file.write(response.text)

print("Saved successfully!")
```

```python
##Parse the HTML

soup = BeautifulSoup(response.text, "html.parser")

print(soup.title.text)
```

```python
##Check whether the product data is present
print(response.text[:1000])
```

```python
###Create a BeautifulSoup object

from bs4 import BeautifulSoup

soup = BeautifulSoup(response.text, "html.parser")
```

```python
##checking page title 

print(soup.title.text)
```

```python
###Check if it's a Shopify store
print("shopify" in response.text.lower())

print("Product" in response.text)

print("price" in response.text.lower())
```

```python
###Now let's inspect all CSS classes
classes = set()

for tag in soup.find_all(True):
    if tag.get("class"):
        classes.update(tag.get("class"))

print(f"Total classes: {len(classes)}")

for cls in sorted(classes):
    print(cls)
```

```python
##These are very typical Shopify product card classes. That means the website is likely rendering product cards in the HTML, which is good news.
```

```python
##Let's find the product cards
products = soup.select(".card-wrapper")

print("Number of product cards:", len(products))
```

```python
##Inspect one product
print(products[0].prettify())
```

```python
###Let's find the product name

print(products[0].get_text(" ", strip=True))

##Find the product link

link = products[0].find("a")
print(link)

print(link.get("href"))

###Find the image

img = products[0].find("img")

print(img)

print(img.get("src"))


##Find the price


prices = products[0].find_all(text=True)

for p in prices:
    if "Rs" in p or "$" in p:
        print(p)
```

```python
###Extract one product


product = products[0]

# Product Name
name = product.select_one("a.full-unstyled-link").get_text(strip=True)

# Product URL
url = "https://lk.spaceylon.com" + product.select_one("a.full-unstyled-link")["href"]

# Image URL
image = "https:" + product.find("img")["src"]

# Price
price = None

for text in product.stripped_strings:
    if text.startswith("Rs"):
        price = text
        break

print("Name :", name)
print("Price:", price)
print("URL  :", url)
print("Image:", image)
```

```python
###Scrape all 18 products

data = []

for product in products:

    # Name
    name = product.select_one("a.full-unstyled-link").get_text(strip=True)

    # URL
    url = "https://lk.spaceylon.com" + product.select_one("a.full-unstyled-link")["href"]

    # Image
    image = "https:" + product.find("img")["src"]

    # Price
    price = None

    for text in product.stripped_strings:
        if text.startswith("Rs"):
            price = text
            break

    data.append({
        "Name": name,
        "Price": price,
        "URL": url,
        "Image": image
    })
```

```python
###Create a DataFrame

df = pd.DataFrame(data)

df.head()
```

```python
###saving
df.to_csv("space_ceylon_products.csv", index=False)

print("CSV saved successfully!")
```
